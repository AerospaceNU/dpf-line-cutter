# Summary of flight

Cherry launched on TRD as the only line cutter on a 60" iris parachute. The reefing lines were rounted through another line cutter casing on the other side of the parachute, but the PCB was missing in that casing because serial coms weren't working the night before. The line cutter appears to have performed perfectly. The apogee shown in the line cutter data is the same as in the easymini data. Disreefing altitudes reported by the line cutter are roughly in agreement with the altitudes of descent speed changes shown in the easymini data. The rocket landed in a large cornfield and we were unable to recover it for over three hours, during which time was there was brief rainstorm. When the line cutter casing was opened to unplug the battery, the nichrome on one of the channels spontaneously burnt through, likely because of water shorting terminals on the mosfet. The day after the launch, the board was dried out and tested. It behaved normally. This test flight confirmed that Cd's determined through car-mounted testing on a 48" parachute translate well to 60" parachutes because the descent speeds shown in the easymini data are close to the target speeds used to set the reefing line lengths.
